```python
import pandas as pd
df=pd.read_excel(r"C:\Users\mdfai\Downloads\Excel Data\Customer Call List.xlsx")
```

```python
df=df.drop_duplicates()
```

```python
df=df.drop(columns="Not_Useful_Column")
```

```python
df["Last_Name"]=df["Last_Name"].str.lstrip("/,...")
```

```python
df["Last_Name"]=df["Last_Name"].str.rstrip("_")
```

```python
df["Paying Customer"]=df["Paying Customer"].replace("Yes","Y")
```

```python
df["Paying Customer"]=df["Paying Customer"].replace("No","N")
```

```python
df["Paying Customer"]=df["Paying Customer"].replace("N/a","")
```

```python
df["Do_Not_Contact"]=df["Do_Not_Contact"].replace("Yes","Y")
```

```python
df["Do_Not_Contact"]=df["Do_Not_Contact"].replace("No","N")
```

```python
df["Phone_Number"]=df["Phone_Number"].replace("N/a","")
```

```python
df["Phone_Number"]=df["Phone_Number"].replace("/","-")
```

```python
df["Phone_Number"]=df["Phone_Number"].replace("|","-")
```

```python
df["Phone_Number"]=df["Phone_Number"].replace("0-9","-")
```

```python
df=df.drop(columns="Phone_Number")
```

```python
df=df.fillna('')
```
