```python
l=df['itemDescription'].value_counts().sort_values(ascending=False)[:10]
```

```python
ax=sns.barplot(l.values,color='cyan')
plt.title("Maximum Sales of Top 10 Products") 
ax.bar_label(ax.containers[0])
plt.show()
```

```python
date=df['Date'].value_counts().sort_values(ascending=False)[:10]
date
```
