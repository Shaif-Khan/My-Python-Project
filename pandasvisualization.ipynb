```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv(r"C:\Users\mdfai\Downloads\Excel Data\icecream.csv")
df.set_index('Flavour')
```

```python
print(plt.style.available)
plt.style.use('fivethirtyeight')
```

```python
df.plot(kind='line',title= 'Ice-Cream Ratings',xlabel='Daily Rating',ylabel='Total Rating')
```

```python
df.plot(kind='line',subplots=True)
```

```python
df.plot.barh(stacked=True,ylabel='Total Rating',xlabel='Daily Basis')
```

```python
df['Texture Rating'].plot.barh(stacked=True,xlabel='Data on Daily Basis',ylabel='Total Texture Rating')
```

```python
df.plot.scatter(x = 'Flavour Rating', y = 'Total Rating',s=50,c='green')
```

```python
df.plot.hist(bins=20)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize=(10,5))
```

```python
df.plot.pie(y='Flavour Rating',figsize=(10,6))
```
